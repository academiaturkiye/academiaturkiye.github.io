**pythonda sınıf yapısı içinde ajanlar oluşturup bu ajanların farklı istatistiksel dağılımlardan çekildiğini gösteren bir kod yazabilir misin?**


Tabii! Python'da sınıf yapısını kullanarak farklı istatistiksel dağılımlardan çekilen ajanları temsil edebiliriz. Aşağıda, her bir ajanı bir sınıf olarak tanımlayarak ve bu ajanların farklı dağılımlardan değerler aldığını gösteren bir örnek vereceğim.

Bu örnekte, `Ajan` isimli bir temel sınıfımız olacak ve bu sınıfın iki alt sınıfı olacak: biri normal dağılımdan veri çekecek, diğeri ise üstel dağılımdan veri çekecek.

İlk olarak, gerekli kütüphaneleri içe aktaracağız:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Sonra, temel `Ajan` sınıfını ve iki alt sınıfı tanımlayacağız:

```python
class Ajan:
    def __init__(self, ad):
        self.ad = ad

    def veri_olustur(self, n):
        raise NotImplementedError("Bu metodu alt sınıflarda implement edin.")
        
    def __str__(self):
        return f"Ajan: {self.ad}"

class NormalAjan(Ajan):
    def __init__(self, ad, ortalama, std):
        super().__init__(ad)
        self.ortalama = ortalama
        self.std = std

    def veri_olustur(self, n):
        return np.random.normal(self.ortalama, self.std, n)

    def __str__(self):
        return f"Normal Ajan: {self.ad}, Ortalama: {self.ortalama}, Std: {self.std}"

class UstelAjan(Ajan):
    def __init__(self, ad, lambd):
        super().__init__(ad)
        self.lambd = lambd

    def veri_olustur(self, n):
        return np.random.exponential(1/self.lambd, n)

    def __str__(self):
        return f"Üstel Ajan: {self.ad}, Lambda: {self.lambd}"
```

Şimdi, bu ajanları kullanarak veri oluşturalım ve bu verileri görselleştirelim:

```python
def veriyi_goster(ajanlar, n=1000):
    plt.figure(figsize=(12, 6))
    
    for ajan in ajanlar:
        veri = ajan.veri_olustur(n)
        plt.figure(figsize=(12, 6))
        sns.histplot(veri, kde=True, stat="density", linewidth=0)
        plt.title(f"{ajan} Verileri")
        plt.xlabel("Değer")
        plt.ylabel("Yoğunluk")
        plt.show()

# Ajanları oluşturma
normal_ajan = NormalAjan("Ajan1", ortalama=0, std=1)
ustel_ajan = UstelAjan("Ajan2", lambd=1)

# Ajanların verilerini görselleştirme
veriyi_goster([normal_ajan, ustel_ajan])
```

### Açıklama:
1. **`Ajan` sınıfı**: Temel bir sınıf olup, alt sınıfların veri oluşturma metodunu implemente etmesini zorunlu kılar.

2. **`NormalAjan` ve `UstelAjan` sınıfları**: `Ajan` sınıfından türemiş ve her biri belirli bir dağılım türünden veri oluşturur. `NormalAjan` sınıfı normal dağılımdan, `UstelAjan` sınıfı ise üstel dağılımdan veri çeker.

3. **`veriyi_goster` fonksiyonu**: Her ajan için veri üretir ve bu verileri seaborn kullanarak histogram ve yoğunluk eğrisi şeklinde görselleştirir.

Bu kod, farklı istatistiksel dağılımlardan veri çeken ajanları nasıl tanımlayabileceğinizi ve bu verileri nasıl görselleştirebileceğinizi gösterir. Her ajan türü, farklı dağılımlardan veri çeker ve bu verileri analiz etmek için uygun araçları sağlar.

````

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````

class Ajan:
    def __init__(self, ad):
        self.ad = ad

    def veri_olustur(self, n):
        raise NotImplementedError("Bu metodu alt sınıflarda implement edin.")

    def __str__(self):
        return f"Ajan: {self.ad}"

class NormalAjan(Ajan):
    def __init__(self, ad, ortalama, std):
        super().__init__(ad)
        self.ortalama = ortalama
        self.std = std

    def veri_olustur(self, n):
        return np.random.normal(self.ortalama, self.std, n)

    def __str__(self):
        return f"Normal Ajan: {self.ad}, Ortalama: {self.ortalama}, Std: {self.std}"

class UstelAjan(Ajan):
    def __init__(self, ad, lambd):
        super().__init__(ad)
        self.lambd = lambd

    def veri_olustur(self, n):
        return np.random.exponential(1/self.lambd, n)

    def __str__(self):
        return f"Üstel Ajan: {self.ad}, Lambda: {self.lambd}"
````

````

def veriyi_goster(ajanlar, n=1000):
    plt.figure(figsize=(12, 6))

    for ajan in ajanlar:
        veri = ajan.veri_olustur(n)
        plt.figure(figsize=(12, 6))
        sns.histplot(veri, kde=True, stat="density", linewidth=0)
        plt.title(f"{ajan} Verileri")
        plt.xlabel("Değer")
        plt.ylabel("Yoğunluk")
        plt.show()

# Ajanları oluşturma
normal_ajan = NormalAjan("Ajan1", ortalama=0, std=1)
ustel_ajan = UstelAjan("Ajan2", lambd=1)

# Ajanların verilerini görselleştirme
veriyi_goster([normal_ajan, ustel_ajan])
````
