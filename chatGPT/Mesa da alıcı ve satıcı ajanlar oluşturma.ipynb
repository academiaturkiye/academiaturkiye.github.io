**Mesa da alıcı ve satıcı ajanlar oluşturma**

Mesada tüccar ajanlar ile ilgili bir örnek verebilir misin?


Elbette! Tüccar ajanları içeren bir örnek, ekonomi veya ticaret simülasyonları gibi karmaşık sistemlerin modellenmesi için yaygın bir kullanım senaryosudur. Burada, iki tür tüccar ajanı içeren basit bir örnek sunacağım: bir `Alıcı` (Buyer) ve bir `Satıcı` (Seller). Bu ajanlar bir pazar yerinde karşılıklı olarak alışveriş yapacaklar.

### Senaryo
- **Alıcı**: Belirli bir bütçeye sahip ve ürünleri satın almaya çalışır.
- **Satıcı**: Ürünleri bir fiyatla sunar ve satmaya çalışır.
- Her iki ajan da belirli bir zaman dilimi içinde hareket eder ve alım-satım işlemleri gerçekleştirir.

### Adımlar

1. **Mesa Kütüphanesini Kurma**:
   ```bash
   pip install mesa
   ```

2. **Ajan ve Modeli Tanımlama**:
   
   İşte bu ajanları ve modeli tanımlayan bir Python kodu:

   ```python
   from mesa import Agent, Model
   from mesa.time import RandomActivation
   from mesa.space import MultiGrid
   from mesa.datacollection import DataCollector
   from mesa.visualization.modules import CanvasGrid
   from mesa.visualization.ModularVisualization import ModularServer
   import random

   class Alici(Agent):
       """Alıcı ajanı."""
       def __init__(self, unique_id, model, budget):
           super().__init__(unique_id, model)
           self.budget = budget

       def step(self):
           # Rastgele bir satıcı ile alışveriş yapma
           seller = self.random.choice(self.model.schedule.agents)
           if isinstance(seller, Satici) and seller.price <= self.budget:
               self.budget -= seller.price
               self.model.grid.remove_agent(seller)
               self.model.schedule.remove(seller)
               print(f"Alici {self.unique_id} {seller.price} ile ürün aldı.")

   class Satici(Agent):
       """Satıcı ajanı."""
       def __init__(self, unique_id, model, price):
           super().__init__(unique_id, model)
           self.price = price

       def step(self):
           # Satıcının fiyatını güncelleme
           self.price = random.uniform(1, 10)
           print(f"Satıcı {self.unique_id} fiyatını {self.price} olarak güncelledi.")

   class PazarModel(Model):
       """Tüccar modelini tanımlar."""
       def __init__(self, N_alicilar, N_saticilar, width, height):
           self.num_agents = N_alicilar + N_saticilar
           self.grid = MultiGrid(width, height, True)
           self.schedule = RandomActivation(self)
           self.running = True

           # Alici ajanları oluştur
           for i in range(N_alicilar):
               a = Alici(i, self, budget=random.uniform(10, 100))
               x = self.random.randrange(self.grid.width)
               y = self.random.randrange(self.grid.height)
               self.schedule.add(a)
               self.grid.place_agent(a, (x, y))

           # Satici ajanları oluştur
           for i in range(N_saticilar):
               s = Satici(i + N_alicilar, self, price=random.uniform(1, 10))
               x = self.random.randrange(self.grid.width)
               y = self.random.randrange(self.grid.height)
               self.schedule.add(s)
               self.grid.place_agent(s, (x, y))

       def step(self):
           self.schedule.step()
   ```

3. **Görselleştirme**:

   Ajanları grid üzerinde göstermek için bir görselleştirici oluşturabilirsiniz:

   ```python
   def agent_portrayal(agent):
       """Ajanların görsel temsili."""
       if isinstance(agent, Alici):
           return {
               "Shape": "circle",
               "Color": "blue",

           }
       elif isinstance(agent, Satici):

               "Shape": "square",

               "Filled": True,
               "w": 0.6
           }
       return {}

   # Grid boyutları
   grid = CanvasGrid(agent_portrayal, 10, 10, 500, 500)

   # Model ve görselleştirme sunucusunu oluştur
   server = ModularServer(
       PazarModel,
       [grid],
       "Pazar Modeli",
       {"N_alicilar": 10, "N_saticilar": 5, "width": 10, "height": 10}
   )

   server.port = 8521
   server.launch()
   ```

### Açıklama

1. **Ajanlar**:
   - **Alıcı**: Belirli bir bütçeye sahip olup, rastgele seçilen bir satıcıdan ürün satın alabilir.
   - **Satıcı**: Rastgele bir fiyat belirler ve bu fiyatla ürün sunar.

2. **Model**:
   - **PazarModel**: Hem alıcı hem de satıcı ajanları oluşturur, onları bir grid üzerine yerleştirir ve simülasyon adımlarını yönetir.

3. **Görselleştirme**:
   - `agent_portrayal` fonksiyonu, alıcıların ve satıcıların grid üzerindeki görsel temsillerini tanımlar.
   - `CanvasGrid`, ajanları 2D grid üzerinde göstermek için kullanılır.

Bu örnek, Mesa kütüphanesi kullanarak alıcı ve satıcı ajanları arasındaki etkileşimleri simüle eden basit bir model sunar. Gerçek dünyada daha karmaşık sistemleri modellemek için ajanların davranışlarını, etkileşimlerini ve pazar koşullarını daha ayrıntılı bir şekilde tanımlamanız gerekebilir.

````
from mesa import Agent, Model
from mesa.time import RandomActivation
from mesa.space import MultiGrid
from mesa.datacollection import DataCollector
from mesa.visualization.modules import CanvasGrid
from mesa.visualization.ModularVisualization import ModularServer
import random

class Alici(Agent):
    """Alıcı ajanı."""
    def __init__(self, unique_id, model, budget):
        super().__init__(unique_id, model)
        self.budget = budget

    def step(self):
        # Rastgele bir satıcı ile alışveriş yapma
        seller = self.random.choice(self.model.schedule.agents)
        if isinstance(seller, Satici) and seller.price <= self.budget:
            self.budget -= seller.price
            self.model.grid.remove_agent(seller)
            self.model.schedule.remove(seller)
            print(f"Alici {self.unique_id} {seller.price} ile ürün aldı.")

class Satici(Agent):
    """Satıcı ajanı."""
    def __init__(self, unique_id, model, price):
        super().__init__(unique_id, model)
        self.price = price

    def step(self):
        # Satıcının fiyatını güncelleme
        self.price = random.uniform(1, 10)
        print(f"Satıcı {self.unique_id} fiyatını {self.price} olarak güncelledi.")

class PazarModel(Model):
    """Tüccar modelini tanımlar."""
    def __init__(self, N_alicilar, N_saticilar, width, height):
        self.num_agents = N_alicilar + N_saticilar
        self.grid = MultiGrid(width, height, True)
        self.schedule = RandomActivation(self)
        self.running = True

        # Alici ajanları oluştur
        for i in range(N_alicilar):
            a = Alici(i, self, budget=random.uniform(10, 100))
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.schedule.add(a)
            self.grid.place_agent(a, (x, y))

        # Satici ajanları oluştur
        for i in range(N_saticilar):
            s = Satici(i + N_alicilar, self, price=random.uniform(1, 10))
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.schedule.add(s)
            self.grid.place_agent(s, (x, y))

    def step(self):
        self.schedule.step()
````

````
def agent_portrayal(agent):
    """Ajanların görsel temsili."""
    if isinstance(agent, Alici):
        return {
            "Shape": "circle",
            "Color": "blue",
            "Filled": True,
            "r": 0.6
        }
    elif isinstance(agent, Satici):
        return {
            "Shape": "square",
            "Color": "red",
            "Filled": True,
            "w": 0.6
        }
    return {}

# Grid boyutları
grid = CanvasGrid(agent_portrayal, 10, 10, 500, 500)

# Model ve görselleştirme sunucusunu oluştur
server = ModularServer(
    PazarModel,
    [grid],
    "Pazar Modeli",
    {"N_alicilar": 10, "N_saticilar": 5, "width": 10, "height": 10}
)

server.port = 8521
server.launch()
````
